# Quantum Phase Estimator

```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
from qiskit import QuantumCircuit, execute, Aer
from qiskit.visualization import plot_histogram
```

```python
circuit = QuantumCircuit(4, 3)

for counting_qubit in range(3):
    circuit.h(counting_qubit)

circuit.x(3)

circuit.draw(output='mpl')
```

```python
repetitions = 1
for counting_qubit in range(3):
    for i in range(repetitions):
        circuit.cu1(math.pi/4, counting_qubit, 3)
    repetitions *= 2
circuit.draw(output='mpl')
```

```python
def qft_dg(n):
    circuit = QuantumCircuit(n, name='qft_dg')
    for qubit in range(n//2):
        circuit.swap(qubit, n-qubit-1)
    for j in range(n):
        for m in range(j):
            circuit.cu1(-math.pi/float(2**(j-m)), m, j)
        circuit.h(j)
    return circuit
```

```python
circuit.append(qft_dg(3).to_instruction(), range(3))
for counting_qubit in range(3):
    circuit.measure(counting_qubit, counting_qubit)
circuit.draw(output='mpl')
```

```python
backend = Aer.get_backend('qasm_simulator')
result = execute(circuit, backend, shots=1024).result()
plot_histogram(result.get_counts())
```
